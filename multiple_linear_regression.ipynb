```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.feature_selection import ExhaustiveFeatureSelector as EFS
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Lasso,Ridge,ElasticNet
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
df = pd.read_csv('Housing.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
X = df.iloc[:,1:]
```

```python
y = df.iloc[:,0:1]
```

```python
label = LabelEncoder()
```

```python
df['mainroad'] = label.fit_transform(df['mainroad'])
```

```python
df['guestroom'] = label.fit_transform(df['guestroom'])
```

```python
df['basement'] = label.fit_transform(df['basement'])
```

```python
df['hotwaterheating'] = label.fit_transform(df['hotwaterheating'])
```

```python
df['airconditioning'] = label.fit_transform(df['airconditioning'])
```

```python
df['prefarea'] = label.fit_transform(df['prefarea'])
```

```python
df['furnishingstatus'] = label.fit_transform(df['furnishingstatus'])
```

```python
df.info()
```

```python
X = df.iloc[:,1:]
```

```python
y = df.iloc[:,0:1]
```

```python
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
lr = LinearRegression()

lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
r2 = r2_score(y_test,y_pred)
```

```python
print('without feature selection r2 score is ',r2)
```

```python
# ridge regression

ridge = Ridge(alpha=0.1)
ridge.fit(X_train,y_train)
```

```python
y_pred_ridge = ridge.predict(X_test)
```

```python
print('r2 score of ridge',r2_score(y_test,y_pred_ridge))
```

```python
# lasso
lasso =Lasso(alpha=0.1)
lasso.fit(X_train,y_train)
```

```python
y_pred_lasso = lasso.predict(X_test)
```

```python
print('r2 score of lasso = ',r2_score(y_test,y_pred_lasso))
```

```python
# elasticnet
elas =ElasticNet(alpha=0.1,l1_ratio=0.5)
elas.fit(X_train,y_train)
```

```python
y_pred_elast = elas.predict(X_test)
```

```python
print('r2 score of elastucnet = ',r2_score(y_test,y_pred_elast))
```

```python
from sklearn.model_selection import GridSearchCV

params = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100]}

grid_ridge = GridSearchCV(Ridge(), params, scoring='r2', cv=5)
grid_ridge.fit(X_train, y_train)
print("Best Ridge alpha:", grid_ridge.best_params_)
print("Best Ridge R²:", grid_ridge.best_score_)
```

```python
params = {
    'alpha': [0.001, 0.01, 0.1, 1, 10],
    'l1_ratio': [0.1, 0.5, 0.7, 0.9]
}

grid_elastic = GridSearchCV(ElasticNet(), params, scoring='r2', cv=5)
grid_elastic.fit(X_train, y_train)
print("Best ElasticNet params:", grid_elastic.best_params_)
print("Best ElasticNet R²:", grid_elastic.best_score_)
```
